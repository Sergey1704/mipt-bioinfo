```python
def reverse(array, index1, index2):
    assert index1 <= index2
    array[index1:index2 + 1] = reversed(array[index1:index2 + 1])
    for i in range(index1, index2 + 1):
        array[i] *= -1
```

```python
def print_sequence(sequence):
    result_str = '('
    for elem in sequence:
        if elem > 0:
            result_str += '+' + str(elem) + ' '
        else:
            result_str += str(elem) + ' '
    result_str = result_str[:-1] + ')'
    print(result_str)
```

```python
filename = 'rosalind_ba6a.txt'
```

```python
with open(filename) as file:
    string = file.readline().rstrip()
    str_seq = string.replace('(', '').replace(')', '').split()
    sequence = [int(number) for number in str_seq]
```

```python
for i in range(len(sequence)):
    if sequence[i] == i+1:
        continue
    try:
        index = sequence.index(i + 1)
    except ValueError:
        index = sequence.index(-i - 1)
        
    reverse(sequence, i, index)
    print_sequence(sequence)
    if sequence[i] < 0:
        reverse(sequence, i, i)
        print_sequence(sequence)
```
