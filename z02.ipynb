```python
data = ''
```

```python
skew = 0
skews = list()
skews.append(skew)

for i in range(len(data)):  
    if data[i] == 'G': skew += 1
    if data[i] == 'C': skew -= 1
    skews.append(skew)

min_skew = min(skews)
min_skews_pos = [i for i in range(len(skews)) if skews[i] == min_skew]

str_answer = ''
for sk in min_skews_pos:
    str_answer += str(sk) + " "
    
print(str_answer)
```
