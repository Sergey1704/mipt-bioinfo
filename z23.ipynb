```python
def get_alignment(string, direction_str, index):
    arrow = '→' if index == 'i' else '↓'
    alignment = ''
    i = 0
    for arr in direction_str:
        if arr == arrow:
            alignment += '-'
        else:
            alignment += string[i]
            i += 1
    return alignment
```

```python
filename = 'rosalind_ba5e.txt'
```

```python
with open(filename) as file:
    str_i = file.readline().rstrip()
    str_j = file.readline().rstrip()
```

```python
n = len(str_i)
m = len(str_j)
sigma = 5
path_matrix = [[0] * (m+1) for i in range(n+1)]
direction_matrix = [['0'] * (m+1) for i in range(n+1)]
```

```python
for i in range(n+1):
    for j in range(m+1):
        if i == 0 and j == 0:
            continue
            
        left = float('-inf')
        up = float('-inf')
        diag = float('-inf')
        
        if i-1 >=0 and j-1 >= 0:
            diag = path_matrix[i-1][j-1] + get_score(str_i[i-1], str_j[j-1])
        if j - 1 >= 0:
            left = path_matrix[i][j-1] - sigma
        if i - 1 >= 0:
            up = path_matrix[i-1][j] - sigma
        
        if left >= up and left >= diag:
            path_matrix[i][j] = left
            direction_matrix[i][j] = 'l'
        elif up >= left and up >= diag:
            path_matrix[i][j] = up
            direction_matrix[i][j] = 'u'
        elif diag >= left and diag >= up:
            path_matrix[i][j] = diag
            direction_matrix[i][j] = 'd'
```

```python
direction_str = ''
i, j = n, m
while (i != 0 or j != 0):
    if direction_matrix[i][j] == 'l':
        j -= 1
        direction_str += '→'
    elif direction_matrix[i][j] == 'u':
        i -= 1
        direction_str += '↓'
    elif direction_matrix[i][j] == 'd':
        i -= 1
        j -= 1
        direction_str += '↘'
        
direction_str = direction_str[len(direction_str)::-1]
```

```python
score = path_matrix[n][m]
alignment_i = get_alignment(str_i, direction_str, 'i')
alignment_j = get_alignment(str_j, direction_str, 'j')
```
